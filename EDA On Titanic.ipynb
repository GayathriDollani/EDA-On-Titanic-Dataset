## Data Cleaning and Pre-processing

### Filling missing 'Age' with median

```python
Titanic['Age'] = Titanic['Age'].fillna(Titanic['Age'].median())
```

```python
Titanic.isnull().sum()
```

### Filling missing 'Embarked' with mode

```python
Titanic['Embarked'] = Titanic['Embarked'].fillna(Titanic['Embarked'].mode()[0])
```

```python
Titanic.isnull().sum()
```

### Dropped Cabin due to Excess missingness

```python
Titanic.drop(columns=['Cabin'], inplace=True)
```

## **Feature Engineering**

### Create FamilySize

```python
Titanic['FamilySize'] = Titanic['SibSp'] + Titanic['Parch'] + 1
```

```python
Titanic['FamilySize']
```

### Survival by FamilySize

```python
sns.countplot(x='Survival', hue='FamilySize', data=Titanic)
plt.title('Survival Rate by Family Size')
plt.show()
```

### Create a 'IsAlone' feature: 1 if Alone, 0 if with family

```python
Titanic['IsAlone'] = np.where(Titanic['FamilySize'] == 1, 1, 0)
```
